# Concentration Inequalities

If you watched a random movies and measured the runtime, do you think that the probability that the runtime is longer than $2$ hours less than 5%? What about 50%. What information about movies would be helpful for you to answer the question? What if we sampled $1,000$ movies, would that make it more or less probable that the average runtime is longer than $2$ hours?

These are the kinds of questions that we can solve with concentration inequalities. We will tackle this problem from a more statistical perspective later in the book, but for now, we look at it purely probablistically. By this, I mean that we assume we know everything about distribution of the random variable, which in this case is runtime.

Let's call the length of a random movie $X$
- What is the sample space that $X$ is mapping from? <details><summary> Click to Show </summary> The sample space is the set of all movies </details>
- Is $X$ a continuous or discrete random variable? <details><summary> Click to Show </summary> It depends! Typically we treat time as continuous, but we also often round to the nearest second or minute. In this problem, we will solve it in both cases.


Let's assume that $X$ is a continuous random variable for now, as the proof will apply to both the discrete and continuous cases

$
\begin{align*}
\mathbb{P}(X \geq a) &= \int_{a}^{\infty} \mathbb{P}(X = a) \\
&\leq \int_{-\infty}^{\infty} \mathbb{P}(X = a) \\
&= \dfrac{1}{a} \int_{-\infty}^{\infty} a \cdot \mathbb{P}(X = a) \\
&= \dfrac{\mathbb{E(X)}}{a}
\end{align*}
$

Cool! We have an upper bound on the probability now! The next question is, whether this upper bound is very good. Let's plug in our numbers, we will assume that the average movie length is $100$ minutes. What is the upper bound on the probability that the average runtime is greater than two hours? <details><summary> Click to Show </summary> $\mathbb{P}(X \geq 120) \leq \dfrac{100}{120} = 5/6$</details>

Is this upper bound very good? Not really, saying that there is less than an 84$ chance that a movies is longer than $2$ hours kind of feels like saying the pyramids were built more than a decade ago, technically correct, but not very helpful.
